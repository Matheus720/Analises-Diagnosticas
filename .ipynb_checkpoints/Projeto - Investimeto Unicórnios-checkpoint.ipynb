```python
# Importar as libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('unicorns till august 2022.csv')
```

```python
investors = df['Investors'].str.split(",",expand=True)
investors = investors.rename(columns = {0:'Investor 1', 1:'Investor 2', 2:'Investor 3', 3:'Investor 4'})
df = df.drop(columns=['Investors'])
df = pd.concat([df, investors],axis=1)
```

- Renomeando colunas para melhor consulta

```python
df.rename( columns={
    'Unnamed: 0' : 'Id',
    'Company' : 'Empresa',
    'Valuation ($B)' : 'Valor ($)',
    'Date Joined' : 'Data de Adesão',
    'Country' : 'Pais',
    'City ' : 'Cidade',
    'Industry': 'Setor',
    }, inplace=True )
```

# Ajustes de Variáveis

```python
df.info()
```

- Nota-se que a coluna 2  está no tipo string, vou converter para numérico, tornando a análise mais precisa
'Valor ($) 936 non-null object' 

```python
df['Valuation ($B)'] = df['Valor ($)'].str.replace(r'$','',regex=True).astype(float)
```

```python
df['Valor ($)'] = pd.to_numeric( df['Valor ($)'].apply( lambda Linha: Linha.replace('$', '') ) )

df.head()
```

```python
# Conversão para Data
df['Data de Adesão'] = pd.to_datetime( df['Data de Adesão'] )
```

```python
# Extrair o Ano e Mes
df['Mes'] = pd.DatetimeIndex( df['Data de Adesão'] ).month
df['Ano'] = pd.DatetimeIndex( df['Data de Adesão'] ).year
```

# Análise Exploratoria

```python
plt.figure( figsize=(15,6) )
plt.title('Analise dos Setores')
plt.bar( df['Setor'].value_counts().index, df['Setor'].value_counts()  )
plt.xticks( rotation=45, ha='right' );
```

```python
plt.figure( figsize=(15,6) )
plt.title('Top 10 Países Geradores de uNICÓRNIOS')
sns.countplot(x = df['Pais'], order = df['Pais'].value_counts().nlargest(10).index)
```

## Unicórnios Brasileiras

```python
brasil = df[df['Pais'] == 'Brazil']
plt.figure(figsize=(25,12))
brasil.plot('Empresa', 'Valor ($)', kind='bar')
```

- Análise Principais Países

# ESTADOS UNIDOS

```python
plt.figure(figsize=(20, 10))
usa_df = df[df['Pais'] == 'United States']
sns.countplot(x = usa_df['Setor'], order = usa_df['Setor'].value_counts().nlargest(5).index)
```

# China

```python
plt.figure(figsize=(20, 10))
usa_df = df[df['Pais'] == 'China']
sns.countplot(x = usa_df['Setor'], order = usa_df['Setor'].value_counts().nlargest(5).index)
```

# India

```python
plt.figure(figsize=(20, 10))
usa_df = df[df['Pais'] == 'India']
sns.countplot(x = usa_df['Setor'], order = usa_df['Setor'].value_counts().nlargest(5).index)
```

### Análise detalhada por ano.

```python
analise_agrupada = df.groupby( by=['Ano'] ).count()['Empresa']

ano = analise_agrupada.index
cont = analise_agrupada.values

plt.figure(figsize=(12,5))
sns.barplot(x = ano, y=cont)
```

- A partir do gráfico acima:
### podemos ver que em 2021 o número de empresas ingressando como startup Unicorn disparou após a recuperação da pandemia.
